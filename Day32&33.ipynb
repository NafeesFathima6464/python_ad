```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic = pd.read_csv("/content/Titanic.csv")
```

```python
titanic.head()
```

```python
titanic['Class'].unique()
```

```python
df=pd.get_dummies(titanic)
df.head()
```

```python
frequent_itemset=apriori(df,min_support=0.1,use_colnames=True)
frequent_itemset
```

```python
x=apriori(df,min_support=0.5,use_colnames=True)
x
```

```python
rules=association_rules(frequent_itemset,metric='lift',min_threshold=0.7)
rules
```
